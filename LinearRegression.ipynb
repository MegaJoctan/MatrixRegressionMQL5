```python
import pandas as pd
import numpy  as np
import sklearn 
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd

datapath = r"NASDAQ_DATA.csv"

data = pd.read_csv(datapath)

print(data.shape)
data.head()
```

```python
import seaborn as sns
datasets = data.iloc[:, 0:]
sns.pairplot(datasets,diag_kind="kde")
```

The Data has to be converted to a numpy array before it can be used to get for the correlation coefficient

```python

y = np.array(data["NASDAQ"])

x_list = ["S&P500","13RSI","50SMA"]

print("Checking the correlation")
for i in x_list: 
    x = np.array(data[i])
    print("\n",i, np.corrcoef(x,y))

```

###### Data has to be reshaped into one column but unkown rows (-1,1)

```python
x = data.drop("NASDAQ",axis=1)

y = data["NASDAQ"].values.reshape(-1,1)


reg = LinearRegression()
reg = reg.fit(x,y)

print("coefficients of our model")
print(reg.coef_)
print(reg.intercept_)
```
